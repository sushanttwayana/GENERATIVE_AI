### Custom Tools

```python
from langchain_core.tools import tool


## Step1 -  Create a function

def multiply(a, b):
    """Multiply two numbers"""
    return a*b

# Step 2- add type hints

def multiply(a: int, b:int) -> int:
    """Multiply two numbers"""
    return a*b

# Step 3 - add tool decorator
@tool
def multiply(a: int, b:int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
result = multiply.invoke({"a":3, "b":5})

print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
### Tool Schema

print(multiply.args_schema.model_json_schema())
```

### Method 2 - Using Structured Tool

```python
from langchain.tools import StructuredTool
from pydantic import BaseModel, Field

class MultiplyInput(BaseModel):
    
    a: float = Field(description="The first number to be added")
    b: float = Field(description="The second number to be added")
```

```python
def multiply_func(a:float, b:float) -> float:
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

### Method 3 - Using BaseTool Class

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic

class MultiplyInput(BaseModel):
    a: int = Field(description="The first number to add")
    b: int = Field(description="The second number to add")
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    ### strict function
    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```

### Toolkit

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
